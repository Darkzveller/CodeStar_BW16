```python
import socket
import time
from tkinter import *
import threading

PORT = 5050
SERVER = socket.gethostbyname(socket.gethostname())
ADDR = (SERVER, PORT)
msg = None
lock = threading.Lock()

class ihm:

    def __init__(self):
        
        self.main_frame= Tk()
        self.main_frame.geometry("1000x540")
        self.main_frame.maxsize(1000,540)
        self.main_frame.minsize(1000,540)

        self.refresh()

    def refresh(self):

        self.nav = Frame(self.main_frame,width=1000, height=50, bg="black")
        self.Modifications = Frame(self.main_frame, width=1000, height = 490, bg="grey94")

        Button(
            self.Modifications,
            bg = "black",
            width = 50,
            height = 20,
            text = "LANCER PAMI",
            fg = "white",
            command = self.send_start_message
        ).place(x = 50, y =80)

        Button(
            self.Modifications,
            bg = "black",
            width = 50,
            height = 20,
            text = "LANCER ETALO",
            fg = "white",
            command = self.send_etalo_message
        ).place(x = 590, y =80)

        self.nav.grid(row = 0, column=0)
        self.Modifications.grid(row = 1, column=0)
    def send_start_message(self):
        global msg
        with lock:
            msg = "START_STAR"

    def send_etalo_message(self):
        global msg
        with lock:
            msg = "ETALO"

def start():
    global msg
    serv = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    serv.bind(ADDR)
    serv.listen()
    print(f"[SERVEUR] En écoute sur {SERVER}:{PORT}")

    conn,addr = serv.accept()

    print(f"[CONNEXION] {addr} connecté.")

    while True:
        with lock:
            if msg != None:
                print(msg)
                conn.send(msg.encode('utf-8'))
                msg = None


if __name__ == "__main__":
    interface = ihm()

    t_wifi = threading.Thread(target=start, daemon=True)
    t_wifi.start()

    interface.main_frame.mainloop()
    
```
